### For future processing we are going to encode our data into numerical form.For that we need to divide our dataset

```python
ecommerce.head(4)
```

```python
ecommerce.shape
```

```python
categ = ecommerce[['city','pincode','online retailers','Easy to use website','Visual appealing','offer',
                                'product info','fast loading website','Reliability','Quickness','payment options','speed delivery',
                               'Privacy','Security of customer','Trustworthiness','online assistance','time to get logged in',
                               'promotion','price','page loading time','mode of payment','delivery period','website/Application design',
                               'Frequent disruption','efficient','Indian online retailer']].copy()
```

```python
categ.shape
```

```python
nominal = ecommerce[['Gender','Age','Since','purchase from past 1yr','internet used','device used','screensize','OS',
                    'Browser_used','channel used for arrival','time taken to purse',
                    'payment option','abondan_shopping','abondant_shapping bag','Content','Information','complete information',
                    'relevant_info','Navgation','speed','userfrndly','payment method','transaction_stipualted time',
                     'empathy towards customer','customer privacy','communication channels','benefits and discounts','enjoyment',
                     'convenients','Return and replace policy','loyalty','quality info','user_satisfaction','net benefits','trust',
                     'product in several category','relavant product info','savings','patronizing','adventure','e-tailer',
                     'gratification','helps','value for money']].copy()
```

```python
nominal.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in nominal:
    nominal[i] = le.fit_transform(nominal[i])
nominal
```

```python
data = pd.concat([nominal,categ],axis=1)
```
